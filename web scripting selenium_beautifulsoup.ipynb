```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import selenium
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
products_list =[]#we will save the dataExtraction to this list.
def DataExtraction():

    page =3 # to increase the scrolling amount in the browser. 
            #you can increase the scrolling amount as much as you want till you reach the end of the products.
            
    driver = webdriver.Chrome(ChromeDriverManager().install()) 
    driver.get("https://www.trendyol.com/sr?mid=655&os=1&pi=3")
    #webdriver opening processes.

    ############################################################
    #Javascript code to browse products and perform cycle actions in the browser is:
    n=0   
    while n<page:

        lastHeight = driver.execute_script("return document.body.scrollHeight")
        i =0
        while i<1:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(3)
            newHeight = driver.execute_script("return document.body.scrollHeight")

            if newHeight == lastHeight:
                break
            else:
                lastHeight = newHeight

            i = i+1


    ################################################################################
    #to reach the link of each product.
        PageSource = driver.page_source
        soup = BeautifulSoup(PageSource,"lxml")
        products= soup.find_all("div",attrs ={"class":"p-card-chldrn-cntnr"})
        for contain in products: 
            linkend= contain.a.get("href")
            Perlink ="https://www.trendyol.com/"
            link=Perlink+linkend
            r1 =requests.get(link)
            soup1 = BeautifulSoup(r1.content,"lxml")
    ################################################################################
    #Using BeautifulSoup, you can reach the title of the data by selecting the class of the data you want
    #no value would be returned if no data was retrieved.

            try:  
                produt_Name=soup1.find("h1",attrs={"class":"pr-new-br"}).text
                #print(produt_Name)
            except:
                produt_Name=None
            try:
                produt_Price=soup1.find("span",attrs={"class":"prc-dsc"}).text
                #print(produt_Price)
            except:
                produt_Price=None
            try:
                produt_Size=soup1.find("div",attrs={"class":"variants"}).text
            except:
                produt_Size=None
            try: 
                produt_Foto=soup1.find("div",attrs={"class":"styles-module_slider__o0fqa"})
                produt_İmage=produt_Foto.img.get('src')
                #print(produt_İmage)
            except:
                produt_İmage=None
            try: 
                produt_favori_count=soup1.find("div",attrs={"class":"fv-dt"}).text
                #print(produt_İmage)
            except:
                produt_favori_count=None
            try: 
                produt_Assessment=soup1.find("a",attrs={"class":"rvw-cnt-tx"}).text
                #print(produt_İmage)
            except:
                produt_Assessment = None
        ##################################################################################
        #Transactions of transferring the retrieved data to the dictionary.
            products_dic ={
                'Product Title':produt_Name,
                'Product price':produt_Price,
                'Prduct size ':produt_Size,
                'produt favori count':produt_favori_count,
                'produt Assessment':produt_Assessment,
                'Product İmage':produt_İmage
            } 

            products_list.append(products_dic) #data listing operations

            n=n+1
        
DataExtraction()    
# Transforming data into DataFrame and saving as csv file.     
df = pd.DataFrame(products_list)
print(df)

df.to_csv('TestData.csv')
```
